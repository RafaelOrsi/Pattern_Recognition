#### Passo 2: Importar bibliotecas

```python
import openai
import requests
import json
```

#### Passo 3: Criar uma chave para a API

```python
API_KEY = "copie sua chave aqui"
```

#### Passo 4: Fazer uma autenticao por meio de um portador de token

```python
headers = {"Authorization": f"Bearer {API_KEY}", "Content-Type": "application/json"}
```

#### Passo 5: Obter o link dos modelos do ChatGPT

```python
link_models = "https://api.openai.com/v1/models"
```

#### Passo 6: Requisitar modelos

```python
request_get = requests.get(link_models, headers=headers)
```

```python
print(request_get)
```

```python
print(request_get.text)
```

#### Passo 7: Escolher um modelo

```python
id_chat = "gpt-4o"
```

#### Passo 8: Obter o link do modelo escolhido

```python
link_chat = "https://api.openai.com/v1/chat/completions"
```

#### Passo 9: Formatar o corpo da mensagem

```python
input_message = "Explique, resumidamente, quais so os critrios de Inovao"
```

```python
body_message = {
    "model": id_chat,
    "messages": [{"role": "user", "content": input_message}]
}
```

#### Passo 10: Converter o corpo da mensagem em json

```python
# JSON  uma forma de representar dados estruturados e transferi-los entre diferentes sistemas.
body_message = json.dumps(body_message)
```

#### Passo 11: Requisitar resposta do modelo ChatGPT

```python
request_post = requests.post(link_chat, headers=headers, data = body_message)
```

```python
print(request_post)
```

```python
print(request_post.text)
```

#### Passo 12: Formatar resposta

```python
resposta = request_post.json()
```

```python
message = resposta["choices"][0]["message"]["content"]
```

```python
print(message)
```
