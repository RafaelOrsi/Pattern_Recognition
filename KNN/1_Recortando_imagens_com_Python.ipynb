# Recortando imagens com Python

```python
# Importando a biblioteca Pillow
from PIL import Image
```

```python
# Carregando a imagem neutra
file = Image.open('original_a.jpeg')
file
```

```python
# Dimensão da imagem
file.size
```

```python
# Selecione as coordenadas de recorte de imagem com crop((x1,y1,x2,y2))
file_crop = file.crop((215,370,650,1020))
file_crop
```

```python
# Dimensão da imagem recortada
file_crop.size
```

```python
# Redimensiona a imagem 
file_resized = file_crop.resize((260,360))
file_resized
```

```python
# Dimensão da imagem
file_resized.size
```

```python
# Salvando a imagem
file_resized.save('minha_foto_a.jpg')
```
