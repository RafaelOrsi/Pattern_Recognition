# Introdução ao KNN (K-Nearest Neighbors) 

O modelo K-Nearest Neighbors (K-Vizinhos Mais Próximos) é um algoritmo de aprendizado de máquina supervisionado usado para classificação e regressão. Ele é baseado no princípio de que amostras semelhantes tendem a pertencer à mesma classe ou ter valores de saída semelhantes. O KNN é um dos métodos mais simples de aprendizado de máquina, mas ainda assim é eficaz em muitos casos.

A ideia central do KNN é que, para classificar ou prever o valor de uma nova amostra, o algoritmo consulta os K exemplos de treinamento mais próximos (vizinhos) da nova amostra e toma uma decisão com base nas classes ou valores desses vizinhos. A proximidade é geralmente medida usando alguma métrica de distância, como a distância euclidiana.



```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
dados = {
    'A': [1,1.5,2,1.5,-1,-0.5,0,-0.5],
    'A2': [-0.5,0,-0.5,-1,1.5,2,1.5,1],
    'B': [1,1.5,1,0.5,-1,-0.5,-1,-1.5],
    'B2': [-1.5,-1,-0.5,-1,0.5,1,1.5,1],
    'y': [1,1,1,1,0,0,0,0],
    'y2': [0,0,0,0,1,1,1,1]    
}
dados = pd.DataFrame(dados)
dados
```

```python
dados_pred = {
    'A': [2.5,1.8,0.5,-1,-1],
    'B': [2,1,0,0,-1.5],
    'A2': [2.5,1.8,0.5,-1,-1],
    'B2': [2,1,0,0,-1.5]    
}
dados_pred = pd.DataFrame(dados_pred)
dados_pred
```

```python
fig, ax = plt.subplots()
ax.scatter(x = dados.A, y=dados.B, c=dados.y, cmap = 'viridis')
ax.scatter(x = dados_pred.A, y=dados_pred.B, c='r', marker = 's')
plt.show()
```

```python
# Criando o classificador
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
# Selecionando os dados de treino
X = dados[['A2','B2']]
# Selecionando o gabarito
y = dados.y2
```

```python
# Selecionando os dados de teste
X_test = dados_pred[['A2','B2']]
```

```python
# Fazendo o treinamento do modelo com a função fit
clf = clf.fit(X,y)
```

```python
# Fazemos a previsão com os dados de teste
y_pred = clf.predict(X_test)
y_pred
```

```python
fig, ax = plt.subplots()
ax.scatter(x = dados.A, y=dados.B, c=dados.y, cmap = 'viridis')
ax.scatter(x = dados_pred.A, y=dados_pred.B, c=y_pred, marker = 's')
plt.show()
```
